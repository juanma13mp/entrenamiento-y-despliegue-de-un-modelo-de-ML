```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('G:/ProjectoScikit/Leads.csv')
print(len(df))
df.head().T
```

```python
replacer = lambda str: str.lower().str.replace(' ', '_')
df.columns = replacer(df.columns.str)
for col in list(df.dtypes[df.dtypes == 'object'].index):
    df[col] = replacer(df[col].str)
df.head().T
```

```python
df["lead_source"].value_counts()
df["totalvisits"].value_counts()
df["page_views_per_visit"].value_counts()
df["last_activity"].value_counts()
df["country"].value_counts()
df["specialization"].value_counts()
df["how_did_you_hear_about_x_education"].value_counts()
df["what_is_your_current_occupation"].value_counts()
df["what_matters_most_to_you_in_choosing_a_course"].value_counts()
df["tags"].value_counts()
df["lead_quality"].value_counts()
df["lead_profile"].value_counts()
df["city"].value_counts()
df["asymmetrique_activity_index"].value_counts()
df["asymmetrique_profile_index"].value_counts()
df["asymmetrique_activity_score"].value_counts()
df["asymmetrique_profile_score"].value_counts()
```

```python
df.tags = df.tags.fillna("unknown")
df.lead_source = df.lead_source.fillna("unknown")
df.city = df.city.fillna("other_cities")
df.lead_source = df.lead_source.fillna("unknown")
df.totalvisits = df.totalvisits.fillna(np.nanmedian(df.totalvisits))
df.page_views_per_visit = df.page_views_per_visit.fillna(0)
df.last_activity = df.last_activity.fillna("unknown")
df.specialization = df.specialization.fillna("unknown")
df.what_is_your_current_occupation = df.what_is_your_current_occupation.fillna("unemployed")
```

```python
df = df.drop(columns='prospect_id')
df = df.drop(columns='lead_number')
df = df.drop(columns='lead_quality')
df = df.drop(columns='lead_profile')
df = df.drop(columns='a_free_copy_of_mastering_the_interview')
df = df.drop(columns='through_recommendations')
df = df.drop(columns='digital_advertisement')
df = df.drop(columns='newspaper_article')
df = df.drop(columns='do_not_call')
df = df.drop(columns='newspaper')
df = df.drop(columns='what_matters_most_to_you_in_choosing_a_course')
df = df.drop(columns='search')
df = df.drop(columns='how_did_you_hear_about_x_education')
df = df.drop(columns='x_education_forums')
df = df.drop(columns='magazine')
df = df.drop(columns='receive_more_updates_about_our_courses')
df = df.drop(columns='update_me_on_supply_chain_content')
df = df.drop(columns='get_updates_on_dm_content')
df = df.drop(columns='i_agree_to_pay_the_amount_through_cheque')
df = df.drop(columns='asymmetrique_activity_index')
df = df.drop(columns='asymmetrique_profile_index')
df = df.drop(columns='asymmetrique_activity_score')
df = df.drop(columns='asymmetrique_profile_score')
df = df.drop(columns='country')
```

```python
categorical = ['lead_origin','lead_source','do_not_email','last_activity','specialization','what_is_your_current_occupation','city','last_notable_activity','tags']

numerical = ['totalvisits','total_time_spent_on_website','page_views_per_visit']

df[categorical].nunique()
```

```python
from sklearn.model_selection import train_test_split
df_train_full, df_test = train_test_split(df, test_size=0.2, random_state=1)

df_train, df_val = train_test_split(df_train_full, test_size=0.33, random_state=1)
y_train = df_train.converted.values
y_val = df_val.converted.values

del df_train['converted']
del df_val['converted']
```

```python
global_mean = df_train_full.converted.mean()
round(global_mean, 3)
```

```python
from sklearn.metrics import mutual_info_score

calculate_mi = lambda col: mutual_info_score(col, df_train_full.converted)

df_mi = df_train_full[categorical].apply(calculate_mi)
df_mi = df_mi.sort_values(ascending=False).to_frame(name='MI')
df_mi
```

```python
train_dict = df_train[categorical + numerical].to_dict(orient='records')
dict(sorted(train_dict[0].items()))
```

```python
from sklearn.feature_extraction import DictVectorizer

dv = DictVectorizer(sparse=False)
dv.fit(train_dict)
```

```python
X_train = dv.transform(train_dict)
X_train[0]
```

```python
dv.get_feature_names_out()
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver='liblinear')
model.fit(X_train, y_train)
```

```python
val_dict = df_val[categorical + numerical].to_dict(orient='records')
X_val = dv.transform(val_dict)
y_pred = model.predict_proba(X_val)
y_pred
```

```python
y_pred = model.predict_proba(X_val)[:, 1]
y_pred
```

```python
churn = y_pred >= 0.5
churn
```

```python
round((y_val == churn).mean(), 3) 
```
